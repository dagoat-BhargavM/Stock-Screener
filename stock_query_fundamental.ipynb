```python
import pandas as pd
import ast
from datetime import date
import warnings
warnings.filterwarnings("ignore")
today = date.today().strftime("%d-%m-%Y")
missing_values = ["NaN", "Nan", " ", ""]
df = pd.read_csv(f"/Users/bhargavmukhami/Desktop/Stock_Prediction_Project/stock_data_fundamental/stock(18-02-2024).csv", na_values=missing_values)
df = df.fillna(0)
```

```python
# Data Cleaning

for i in range(len(df)):
    if isinstance(df["Market Cap"][i], str):
        df["Market Cap"][i] = float(df["Market Cap"][i].replace(",", ""))
    if isinstance(df["Compounded Sales Growth"][i], str):
        if df["Compounded Sales Growth"][i].rstrip("%") != "":
            df["Compounded Sales Growth"][i] = float(df["Compounded Sales Growth"][i].rstrip("%"))
    if isinstance(df["Compounded Profit Growth"][i], str):
        if df["Compounded Profit Growth"][i].rstrip("%") != "":
            df["Compounded Profit Growth"][i] = float(df["Compounded Profit Growth"][i].rstrip("%"))
    if len(ast.literal_eval(df["Promoter holding"][i])) > 0:
        l = []
        for j in ast.literal_eval(df["Promoter holding"][i]):
            if j.rstrip("%") != "":
                l.append(float(j.rstrip("%")))
        df["Promoter holding"][i] = l
    else:
        df["Promoter holding"][i] = []
    if len(ast.literal_eval(df["FIIs"][i])) > 0:
        l = []
        for j in ast.literal_eval(df["FIIs"][i]):
            if j.rstrip("%") != "":
                l.append(float(j.rstrip("%")))
        df["FIIs"][i] = l
    else:
        df["FIIs"][i] = []
    if len(ast.literal_eval(df["DIIs"][i])) > 0:
        l = []
        for j in ast.literal_eval(df["DIIs"][i]):
            if j.rstrip("%") != "":
                l.append(float(j.rstrip("%")))
        df["DIIs"][i] = l
    else:
        df["DIIs"][i] = []
    if len(ast.literal_eval(df["Public"][i])) > 0:
        l = []
        for j in ast.literal_eval(df["Public"][i]):
            if j.rstrip("%") != "":
                l.append(float(j.rstrip("%")))
        df["Public"][i] = l
    else:
        df["Public"][i] = []
    if len(ast.literal_eval(df["np_quarterly"][i])) > 0:
        l = []
        for j in ast.literal_eval(df["np_quarterly"][i]):
            l.append(float(j.replace(",", "")))
        df["np_quarterly"][i] = l
    else:
        df["np_quarterly"][i] = []
    if len(ast.literal_eval(df["np_yearly"][i])) > 0:
        l = []
        for j in ast.literal_eval(df["np_yearly"][i]):
            l.append(float(j.replace(",", "")))
        df["np_yearly"][i] = l
    else:
        df["np_yearly"][i] = []
    df["ROE"][i] = float(str(df["ROE"][i]).replace(",", ""))
    df["ROCE"][i] = float(df["ROCE"][i])
    df["Stock P/E"][i] = float(str(df["Stock P/E"][i]).replace(",", ""))
df
```

```python
# Query

final = []
for i in range(len(df)):
    if (5000 < df["Market Cap"][i] < 20000):
        if (5 < df["Stock P/E"][i] < 30):
            if (len(df["Promoter holding"][i]) > 0 or len(df["Promoter holding"][i]) == 0):
                if (len(df["Promoter holding"][i]) == 0 or df["Promoter holding"][i][-1] > 50):
                    if (df["ROE"][i] > df["Stock P/E"][i]):
                        if (df["ROCE"][i] > df["Stock P/E"][i]):
                            if (len(df["np_quarterly"][i]) > 1):
                                if (df["np_quarterly"][i][-1] > df["np_quarterly"][i][-2]):
                                    if (len(df["np_yearly"][i]) > 1):
                                        if (df["np_yearly"][i][-1] > df["np_yearly"][i][-2]):
                                            final.append(df["SYMBOL"][i])
final
```
